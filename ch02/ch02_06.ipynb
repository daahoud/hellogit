# 02-06 if文で条件分岐をする

## 「条件」をプログラムにする

```python
# 2・6・1 ： 平成何年かを計算する
year = 2006
wareki_year = year - 1988
print(wareki_year)
```

```python
# 2・6・2 ： if文を使って平成と令和の年号を計算する
year = 2020
wareki_year = year - 1988
if year > 2018:  # yearが2018より大きかったら
    wareki_year = year - 2018
print(wareki_year)
```

```python
# 2・6・3 ： プログラムの実行を止めながら確認する
year = 2020
wareki_year = year - 1988
input("エンターを押してください")  # プログラムの実行を止める
if year > 2018:  # yearが2018より大きかったら
    input("ifの中(1)")
    wareki_year = year - 2018
    input("ifの中(2)")
print(wareki_year)
```

## 条件を指定する条件式

```python
# 2・6・4 ： 和暦(昭和，平成，令和)
year = 2020
era_name = "昭和"
wareki_year = year - 1925
if year > 1989:
    era_name = "平成"
    wareki_year = year - 1988
if year > 2019:
    era_name = "令和"
    wareki_year = year - 2018
wareki = era_name+f"{wareki_year}年"
print(wareki)
```

## 条件のプログラムを分割して動かす

## elseとelif

```python
# 2・6・8 ： 和暦その2
year = 2020
era_name = "昭和"
wareki_year = year - 1925
if year < 2019:
    era_name = "平成"
    wareki_year = year - 1988
else:
    era_name = "令和"
    wareki_year = year - 2018
wareki = f"{era_name}{wareki_year}年"
print(wareki)
```

```python
# 2・6・9 ： 和暦その3
year = 1989
era_name = "昭和"
wareki_year = year - 1925
if 1989 <= year < 2019:
    era_name = "平成"
    wareki_year = year - 1988
elif 2019 <= year:
    era_name = "令和"
    wareki_year = year - 2018
wareki = f"{era_name}{wareki_year}年"
print(wareki)
```
