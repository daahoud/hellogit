# 02-05 文字列を使う

## データとしての文字列

```python
# 2・5・1
print(10)
```

```python
# 2・5・2 ： 文字列を表示する
print("みんなのPython")
```

```python
# 2・5・3 ： 文字列を変数に代入する
book_title = "みんなのPython"
print(book_title)
```

```python
# 2・5・4 ： print()に複数の引数を与える
temperature = 23
print("ただいまの気温(℃):", temperature)
```

## 「リテラル」と文字列の足し算、掛け算

```python
# 2・5・5 : 文字列の足し算
era_name = "令和"   # 元号名
year_name = "7年"  # 何年か
wareki = era_name + year_name
print(wareki)
```

```python
# 2・5・6 : 文字列の掛け算
delimiter = "-" * 20  # 区切り文字列
print(delimiter)
```

## 数値を文字列に変換する

```python
# 2・5・7
year = 2025   # 西暦
wareki_year = year - 2018
wareki_year   # 結果を確認
```

```python
# 2・5・8
wareki = era_name + wareki_year + "年"
```

```python
# 2・5・9 ： 関数で数値を文字列に変換
wareki_str = str(wareki_year)
```

```python
# 2・5・10
wareki = era_name + wareki_str + "年"
print(wareki)
```

## 「f文字列」を使う

```python
# 2・5・11 ： f文字列を使う
wareki = f"令和{wareki_year}年"
print(wareki)
```

## 「小実験をしてまとめる」を繰り返す

```python
# 2・5・12
year = 2025        # 西暦
wareki_year = year - 2018
wareki = f"令和{wareki_year}年"
print(wareki)
```
